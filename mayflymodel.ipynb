# A Report on the Mayfly Model 

## Introduction
In this project, we are exploring the behavior of a simple mathematical model known as the Mayfly model. This model is used to predict changes in populations, specifically the population of mayflies, a type of insect.

The Mayfly model is based on the assumption that the size of the population in a given year depends only on the size of the population in the previous year. The model considers factors such as the growth rate of the population and the maximum population size that the environment can support.Our goal is to analyze and visualize how the population changes over time for different values of the growth rate and initial population. We will simulate the model over a certain number of years and plot the population for each year.

By observing the dynamics of the model for different parameters, we aim to gain insights into the factors that influence population growth and the complex behaviors that can emerge from simple mathematical rules.This project is a great opportunity to learn about mathematical modeling, population dynamics, and data visualization. It also provides a practical application of programming in Python, particularly the use of loops, functions, and plotting libraries.

## Initial Analysis and Visualization of the MayFly Model 

```python
# Importing Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
#The Mayfly Model Function
def mayfly_model(b, y0, years):
    population = [y0]
    for _ in range(years):
        y = b * (1 - population[-1]) * population[-1]
        population.append(y)
    return population
```

### The b and y0 parameters for initial visualization

```python
# growth rate
b = 2.5  

# initial population
y0 = 0.1  

# number of years to simulate
years = 20  

population = mayfly_model(b, y0, years)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(population, 'o-')
plt.title('Mayfly population over time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

We used the given code to simulate mayfly population dynamics over a 20-year period, with a fixed growth rate ('b') and initial population ('y0'). Based on this simulation, we may make the following observations and formulate hypotheses:


1. **Exponential Growth**: Mayfly populations grow rapidly at first, following an exponential pattern. This is obvious in the long-term trend of increasing population size.

2. **Population Stability**: Over time, population growth begins to stabilize. Eventually, the population size may stabilize or vary around a specific figure. This suggests that the environment's carrying capacity could limit population increase.

3. **Parameter Sensitivity**: We expect that modifying the growth rate ('b') and initial population ('y0') values will have a major impact on population dynamics. For example, increasing the growth rate or initial population may result in quicker or larger population increase.


4. **Threshold Effects**: There may be particular growth rate or initial population thresholds that cause population dynamics to change dramatically. We can test this idea by changing the settings and watching the impact on population increase.


5. **Long-term Stability**: We can determine whether population dynamics eventually stabilize into a steady state or continue to fluctuate indefinitely. This can provide information about the long-term viability of the mayfly population under various parameter settings.

6. **Comparative Analysis**: We can evaluate population dynamics across different parameter combinations to determine which combinations provide more stable or desired outcomes in terms of population size and growth rate.





# Testing out Observations 

Using different values, the observations were tested out to test their behavior

## Observation 1: Exponential Growth


```python
# Define parameters
b = 2.5  # Growth rate
y0 = 0.1  # Initial population
years = 20  # Number of years to simulate

# Simulate population dynamics
population = mayfly_model(b, y0, years)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(population, 'o-')
plt.title('Exponential Growth of Mayfly Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

## Conclusion 
The plot illustrates the rapid increase in population size over time, indicating exponential growth in the early stages of the simulation.

## Observation 2: Population Stability


```python
# Define parameters
b = 2.5  # Growth rate
y0 = 0.1  # Initial population
years = 100  # Number of years to simulate

# Simulate population dynamics
population = mayfly_model(b, y0, years)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(population, 'o-')
plt.title('Population Stability of Mayfly Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

## Conclusion: 
    The plot shows that after a period of exponential growth, the population stabilizes around a certain level, indicating that the environment's carrying capacity limits population growth.



## Observation 3: Sensitivity to Parameters


```python
# Define parameters
b_values = [1, 2, 3]  # Growth rate values to test
y0 = 0.1  # Initial population
years = 20  # Number of years to simulate

# Plot the results
plt.figure(figsize=(10, 6))
for b in b_values:
    population = mayfly_model(b, y0, years)
    plt.plot(population, 'o-', label=f"b={b}")
plt.title('Sensitivity of Mayfly Population to Growth Rate')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion: 
    The plot shows that higher growth rates lead to faster population growth, indicating sensitivity to changes in the growth rate parameter.



## Observation 4: Threshold Effects


```python
# Define parameters
b_values = [2.5, 3, 4]  # Growth rate values to test
y0 = 0.1  # Initial population
years = 20  # Number of years to simulate

# Plot the results
plt.figure(figsize=(10, 6))
for b in b_values:
    population = mayfly_model(b, y0, years)
    plt.plot(population, 'o-', label=f"b={b}")
    
plt.title('Threshold Effects in Mayfly Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusion:
The plot demonstrates that beyond a certain growth rate threshold, the population dynamics change drastically, indicating threshold effects in the model.



## Observation 5: Long-term Stability


```python
# Define parameters
b = 2.5  # Growth rate
y0 = 0.1  # Initial population
years = 200  # Number of years to simulate

# Simulate population dynamics
population = mayfly_model(b, y0, years)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(population, 'o-')
plt.title('Long-term Stability of Mayfly Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

## Conclusion: 
    The plot demonstrates that the population dynamics eventually stabilize to a steady state after a prolonged period, indicating long-term stability under the specified parameter settings.



## Observation 6: Comparative Analysis


```python
# Define parameters
b_values = [2, 2.5, 3]  # Growth rate values to test
y0_values = [0.1, 0.3, 0.5]  # Initial population values to test
years = 20  # Number of years to simulate

# Plot the results
plt.figure(figsize=(12, 8))
for b in b_values:
    for y0 in y0_values:
        population = mayfly_model(b, y0, years)
        plt.plot(population, 'o-', label=f"b={b}, y0={y0}")
plt.title('Comparative Analysis of Mayfly Population Dynamics')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

## Conclusion:
     The plot allows for a comparative analysis of population dynamics under different parameter combinations, providing insights into which combinations lead to more stable or desirable outcomes
